```python
##renomear os títulos das colunas

ano2023.columns = renomear_cp_23
ano2022.columns = renomear_cp_22
ano2021.columns = renomear_cp_21
ano2020.columns = renomear_cp_20
ano2019.columns = renomear_cp_19
ano2018.columns = renomear_cp_18
```

Outra etapa importante deste processo é a edição do campo string, especialmente nas coluna que servirão de elo de ligação entre diferentes tabelas, como é caso a coluna UC (a coluna Unidade de Conservação já renomeada). Primeiro esta coluna terá o string formatada para que o texto fique todo em caixa alta, depois as colunas string serão convertidos para ASCII para evitar problema na exportação do arquivo CSV.  

```python
#Formatando o texto para caixa alta

ano2023['UC'] = ano2023['UC'].str.upper()
ano2022['UC'] = ano2022['UC'].str.upper()
ano2021['UC'] = ano2021['UC'].str.upper()
ano2020['UC'] = ano2020['UC'].str.upper()
ano2019['UC'] = ano2019['UC'].str.upper()
ano2018['UC'] = ano2018['UC'].str.upper()
ano2021['Bioma'] = ano2021['Bioma'].str.upper()
ano2020['Bioma'] = ano2020['Bioma'].str.upper()
ano2019['Bioma'] = ano2019['Bioma'].str.upper()
ano2018['Bioma'] = ano2018['Bioma'].str.upper()
```

```python
#Eliminando caracteres especiais e acentuação (conversão para ASCII)

ano2023['UC'] = ano2023['UC'].apply(unidecode)
ano2022['UC'] = ano2022['UC'].apply(unidecode)
ano2021['UC'] = ano2021['UC'].apply(unidecode)
ano2020['UC'] = ano2020['UC'].apply(unidecode)
ano2019['UC'] = ano2019['UC'].apply(unidecode)
ano2018['UC'] = ano2018['UC'].apply(unidecode)
ano2023['NGI'] = ano2023['NGI'].apply(unidecode)
ano2021['Bioma'] = ano2021['Bioma'].apply(unidecode)
ano2020['Bioma'] = ano2020['Bioma'].apply(unidecode)
ano2019['Bioma'] = ano2019['Bioma'].apply(unidecode)
ano2018['Bioma'] = ano2018['Bioma'].apply(unidecode)
```

```python
# Exportar os dataframe em csv

ano2023.to_csv('area_atingida_2023.csv')
ano2022.to_csv('area_atingida_2022.csv')
ano2021.to_csv('area_atingida_2021.csv')
ano2020.to_csv('area_atingida_2020.csv')
ano2019.to_csv('area_atingida_2019.csv')
ano2018.to_csv('area_atingida_2018.csv')
```
